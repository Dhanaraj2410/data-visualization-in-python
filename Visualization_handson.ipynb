# Create scatterplot for the following data, make sure all the plots appear in the same plane with x-label and y-label evenly spaced.

a. Create scatter plots with between (mpg-hp, mpg-weight, hp-mpg, hp-time-to-60) with respect to each brand.

```python
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)
sns.scatterplot(df,x="mpg",y="hp",hue="brand")
plt.title("Mileage vs Power")
plt.subplot(2,2,2)
sns.scatterplot(df,x="mpg",y="weightlbs",hue="brand")
plt.title("Mileage vs Weight of the Cars")
plt.subplot(2,2,3)
sns.scatterplot(df,x="hp",y="mpg",hue="brand")
plt.title("Power vs Mileage")
plt.subplot(2,2,4)
sns.scatterplot(df,x="hp",y="time-to-60",hue="brand")
plt.title("Power vs time")
plt.show()

```

# 1.b) create a bar plot for the following make sure all the plots appear in the same plane with x-label and y-label evenly spaced.

*  create a bar plot that show the visual representation of hp, mpg, weight,time-to-60 with respect to to the number of cylinders of the brands.

```python
fig = plt.subplots(2,2,figsize=(8,10))
plt.subplot(2,2,1)
sns.barplot(df, y='hp', x='cylinders', hue='brand', palette='husl')
plt.subplot(2,2,2)
sns.barplot(df, y='mpg', x='cylinders', hue='brand', palette='hls')
plt.subplot(2,2,3)
sns.barplot(df, y='weightlbs', x='cylinders', hue='brand', palette='viridis')
plt.subplot(2,2,4)
sns.barplot(df, y='time-to-60', x='cylinders', hue='brand', palette='magma')
```

# create a box plot for the following make sure all the plots appear in the same plane with x-label and y-label evenly spaced.



*   create a bar plot that show the visual representation of hp, mpg, weight,time-to-60 with respect to to the number of cylinders of the brands.



```python
fig = plt.subplots(2,2,figsize=(8,10))
plt.subplot(2,2,1)
sns.boxplot(df, y='hp', x='cylinders', hue='brand')
plt.subplot(2,2,2)
sns.boxplot(df, y='mpg', x='cylinders', hue='brand')
plt.subplot(2,2,3)
sns.boxplot(df, y='weightlbs', x='cylinders', hue='brand')
plt.subplot(2,2,4)
sns.boxplot(df, y='time-to-60', x='cylinders', hue='brand')
```
