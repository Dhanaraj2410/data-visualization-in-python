```python
#adding titles and labels
plt.title('title of the plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()# this will show the graph only.
```

# line plot: display data points connected by stright line .
shows trends over the time.
can plot one or more mupliple lines for the diffrenet dataset.

```python
x1=[1,2,3,4,5]
y1=[2,3,4,5,6]

x2=[1,3,8,4,9]
y2=[2,13,4,5,9]


#marker: *,o,D,x,^,<,>,1,2,3,4
#edge colour:mec
#face colours: mfc
#line style:'-','--','-.'

plt.plot(x1,y1,x2,y2,marker='*',mec='black',mfc='blue',linestyle='-.',color='red',linewidth=3)
plt.title('title of the line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# Bar plot:used to compare categories.
X:catgories(name, places,months..)
y:(numerical/categical both)

```python
categories=['dev','ganesh','nikhil','rudra']
marks=[80,30,50,60]
plt.bar(categories,marks,color='pink')

plt.show()
```

# Histogram
frequency distribution of numeric data.


```python
x=[55,65,75,85,95,100,105]
plt.hist(x,color='red',edgecolor='black')

plt.xlabel('score')
plt.ylabel('number of the student')
plt.title('histogram plot')
plt.show()
```

# Scatter plot: display datapoints as inividual data markers.
use: correlation netween variable.

```python
study_hours=[1,2,3,4,5,6,0,8,9]
exam_score=[10,20,30,40,50,100,70,80,90]

plt.scatter(study_hours,exam_score,color='black')
plt.show()
```

# Pie Chat: display data as slices of a circle. shows propertion of each category relative to the whole.


```python
fruites=['apple','banana','cherries','dates']
sizes=[25,35,20,20]
colours_fru=['red','yellow','darkred','brown']

plt.pie(sizes,labels=fruites,colors=colours_fru,autopct='%1.1f%%')

plt.show()
```

# Seaborn

Used for simpler plots, it has simpler syntax,less cutomization.


```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.scatterplot(df,x='total_bill',y='tip')
plt.show()
```

```python
sns.scatterplot(df,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
df.columns
```

```python
sns.lineplot(data=df,x='size',y='total_bill')
plt.title('line plot without hue')
plt.show()
```

```python
sns.lineplot(data=df,x='size',y='total_bill',hue='day')
plt.title('line plot without hue')
plt.show()
```

```python
df.columns
```

```python
#without hue
sns.barplot(data=df,x='day',y='total_bill')
plt.show()



#without hue
sns.barplot(data=df,x='day',y='total_bill', hue='smoker')
plt.show()
```

# Box plot


```python
sns.boxplot(data=df,x='day',y='total_bill')
plt.show()


sns.boxplot(data=df,x='day',y='total_bill',hue='sex')
plt.show()
```

# Violin Plot


```python
sns.violinplot(data=df,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
#histogram
sns.histplot(data=df,x='total_bill',hue='sex',kde=True)
plt.show()
```

```python
#KDE
sns.kdeplot(data=df,x='total_bill')
plt.show()
```

# Heat Map

This will give correlation between numerical values.

+1-->perfect positive correlation-->both increases-->more hours studied-->higher marks

0--> no correlation-->shoe size vs marks

-1-->perfect negative correlation-->one increases , other decreases-->more gaming, lower exam scores.

```python
#heatmap
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='Reds')
plt.show()
```

```python
#pair plot
sns.pairplot(data=df)
plt.show()


#pair plot
sns.pairplot(data=df,hue='sex')
plt.show()
```
